```python
# Step 1: Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Load your data (assuming you've already loaded the DataFrame 'df')
# Let's assume df is your DataFrame, with the last column as the target

# Separate features and target variable
X = df.iloc[:, 1:-1]  # Features from the second column to the penultimate one
y = df.iloc[:, -1]  # The last column as the target variable

# Step 3: Encode target variable if needed (optional if already 1s and 2s)
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)  # Transforms the target to 0 and 1

# Step 4: Calculate correlation coefficients between features and target
correlation_values = X.apply(lambda col: col.corr(pd.Series(y)))

# Step 5: Calculate statistics (mean, median, standard deviation)
mean_corr = np.mean(correlation_values)
median_corr = np.median(correlation_values)
std_corr = np.std(correlation_values)

# Step 6: Plotting correlation coefficients
plt.figure(figsize=(10, 6))
plt.barh(correlation_values.index, correlation_values.values, color='skyblue')

# Add lines for mean, median, and standard deviation
plt.axvline(mean_corr, color='red', linestyle='--', label=f'Mean: {mean_corr:.2f}')
plt.axvline(median_corr, color='green', linestyle='--', label=f'Median: {median_corr:.2f}')

# Add lines for std deviations
plt.axvline(mean_corr + std_corr, color='orange', linestyle='--', label=f'Mean + 1 STD: {mean_corr + std_corr:.2f}')
plt.axvline(mean_corr - std_corr, color='orange', linestyle='--', label=f'Mean - 1 STD: {mean_corr - std_corr:.2f}')

# Add lines for 0.5, 1.5, and 2 standard deviations
plt.axvline(mean_corr + 0.5 * std_corr, color='purple', linestyle='--', label=f'Mean + 0.5 STD: {mean_corr + 0.5 * std_corr:.2f}')
plt.axvline(mean_corr - 0.5 * std_corr, color='purple', linestyle='--', label=f'Mean - 0.5 STD: {mean_corr - 0.5 * std_corr:.2f}')

plt.axvline(mean_corr + 1.5 * std_corr, color='blue', linestyle='--', label=f'Mean + 1.5 STD: {mean_corr + 1.5 * std_corr:.2f}')
plt.axvline(mean_corr - 1.5 * std_corr, color='blue', linestyle='--', label=f'Mean - 1.5 STD: {mean_corr - 1.5 * std_corr:.2f}')

plt.axvline(mean_corr + 2 * std_corr, color='brown', linestyle='--', label=f'Mean + 2 STD: {mean_corr + 2 * std_corr:.2f}')
plt.axvline(mean_corr - 2 * std_corr, color='brown', linestyle='--', label=f'Mean - 2 STD: {mean_corr - 2 * std_corr:.2f}')

# Add labels and title
plt.xlabel('Correlation Coefficients')
plt.title('Correlation Coefficients with Mean, Median, and Various STD Intervals')
plt.legend()
plt.show()
```
